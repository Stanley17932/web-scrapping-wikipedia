```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Kenya'
page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find('table',class_='wikitable sortable')
# <table class= ''>
# <caption>
```

```python
soup.find_all('th')
```

```python
company_titles = soup.find_all('th')
```

```python
company_table_titles = [title.text.strip() for title in company_titles]
```

```python
print(company_table_titles)
```

```python
import re

text = "vteList of companies of Africa"
updated_text = re.sub(r'^vte', '', text)
print(updated_text)
```

```python
company_table_titles.append(updated_text)
```

```python
company_table_titles
```

```python
# Create a new list without items starting with 'vte'
new_company_list = [item for item in company_table_titles if not item.startswith('vte')]
```

```python
print(new_company_list)
```

```python
#find the index of Notes
index_notes = new_company_list.index('Notes')
```

```python
#remove 'List of companies of Africa' from its current position
new_company_list.remove('List of companies of Africa')
```

```python
#insert 'List f companies of Africa' after 'Notes'
new_company_list.insert(index_notes+1,'List of companies of Africa')
```
